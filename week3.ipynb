### DAY 3
#### UNIVARIATE PLOTS 
##### Focus on analyzing and visualizing a single variable at a time.
##### Key goals include describing its distribution, central point (like mean or median), and variability (such as range or standard deviation).

##### (a)Histogram

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,color='skyblue',edgecolor='white')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

##### (b)Boxplot

```python
plt.boxplot(data,vert=False) # parameter vert controls the orinentation of the boxplot in matplotlib
plt.xlabel('value')
plt.show()
```

#### core elements of a boxplot
##### Box: The central box shows the interquartile range (IQR), which contains the middle 50% of the data—spanning from Q1 (25th percentile) to Q3 (75th percentile).
##### Median Line: A line inside the box marks the dataset’s median (Q2), highlighting the center of the distribution.
##### Whiskers: Lines (the “whiskers”) extend outward from the box to the minimum and maximum values within 1.5 times the IQR, summarizing typical data spread.
##### Outliers: Data points beyond the whiskers are plotted individually and indicate potential extreme values.
##### Min/Max: The smallest and largest values inside the whisker limits are shown as endpoints of the whiskers.

```python
plt.boxplot([-16,-10,1,2,3,4,5,30,43])
plt.show()
```

#### (e)Pairplot
##### A pairplot in Seaborn is a handy visualization tool that creates a grid of plots showing relationships between each pair of variables in a dataset.
##### Pairplot automatically creates:
##### Scatterplots for every pair of numerical columns (off-diagonal).
##### Distribution plots (like histograms or KDE) for each single column (diagonal).


```python
sns.pairplot(iris)
plt.show()
```

##### No hue parameter: (above plotting)
##### Diagonal plots: By default, histograms (univariate distribution using plt.hist).
##### Off-diagonal: Scatter plots.

```python
q=sns.pairplot(iris,hue='species')
plt.show()
```

##### With hue parameter: (above plotting)
##### Diagonal plots: Switch from histogram to univariate KDE plot.
##### Off-diagonal: Scatter plots, with points colored according to hue

```python
sns.pairplot(iris,hue='species',diag_kind='hist',palette='husl')
plt.show()
```

#### (Jointplot)
##### Jointplot (in seaborn) is a visual tool used to show the relationship between two variables, and display each variable’s distribution at the same time.

```python
sns.jointplot(x='tip',y='total_bill',data=tips)
plt.style.use('default')
```

##### Center: A plot (scatter, histogram, KDE, or regression) displaying how the two variables relate to each other.
##### Top & Right Margins: Two extra plots showing the distribution for each individual variable.
##### Top = x-axis variable’s distribution (like a histogram).
##### Right = y-axis variable’s distribution.

### Overlaying two plots:
##### 

```python
sns.boxplot(x='petal_length', y='petal_width', data=iris)          # Plot the boxplot first
sns.stripplot(x='petal_length', y='petal_width', data=iris,        # Overlay the strip plot
              color='red', alpha=0.5,jitter=True)        # Jitter avoids overlap, red dots, slightly transparent
plt.xticks(rotation=90)
plt.show()
```
