### MATPLOTLIB
#### Matplotlib is a widely used Python library for creating static, animated, and interactive visualizations, such as graphs and charts

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
```

```python
plt.plot(x,y)
plt.show
```

```python
plt.plot(x,y,color='blue',marker='D',ms=8,markerfacecolor='red',linestyle='--')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('simple line plot using matplotlib')
plt.show
```

### figure v/s axes
#### figure
##### *The Figure is the top-level container for all plot elements in Matplotlib; it acts as the canvas that holds everything.
##### *A single Figure can contain one or multiple Axes objects, as well as titles, legends, and other annotations.

#### axes
##### *Axes are the actual regions within a Figure where data is plotted, sometimes called subplots
##### *Each Axes manages its own x- and y-axis (or z for 3D), labels, title, tick marks, and the graphical details of the data

```python
fig,ax=plt.subplots()
ax.set_title('Simple line plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
plt.show
```

```python
fig, ax = plt.subplots()  # This creates a Figure and one Axes (the plot area)
ax.plot([1, 2, 3], [1, 4, 9])  # Plot a simple line
plt.show()
```

```python
x=[0,1,2,3,4]
y1=[i for i in x]
y2=[i*2 for i in x]
plt.plot(x,y1,label='linear')
plt.plot(y1,y2,label='quadratic')
plt.show
```

```python
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]
plt.plot(x,y1,color='red',marker='H',ms=8,label='Double')
plt.plot(x,y2,color='blue',marker='^',label='odd')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show
```

#### BAR PLOT
##### Displays rectangular bars to compare data values across categories.
##### Can show vertical or horizontal bars.
##### Created with plt.bar() or plt.barh().

```python
categories= ['A','B','C','D']
values=[10,20,30,40]
plt.bar(categories,values,color='orange')
plt.title('BAR CHART')
plt.show
```

```python
months=['jan','feb','mar','apr','may']
sales=[200,100,120,432,99]
plt.bar(months,sales,color='yellow'
        )
plt.title('Bar chart ')
plt.show()
```

#### HISTOGRAM PLOT
##### Shows the distribution of continuous data by grouping it into bins.
##### Created with plt.hist()

```python
import numpy as np
data=np.random.randn(100)
print(data)
plt.hist(data,bins=10,edgecolor='black')
plt.title('histogram of random data')
plt.xlabel('value range')
plt.ylabel('frequency')
plt.show
```

```python
plt.hist(data,bins=20,color='orange',edgecolor='red')
plt.xlabel('value range')
plt.ylabel('frequency')
plt.title('histogram with 20 bins')
plt.show()
```

```python
plt.hist(data,bins=15,color='coral',edgecolor='white',alpha=0.5)
plt.show
```

##### alpha controls the transarency

```python
dt1=np.random.normal(0,1,100)
dt2=np.random.normal(2,1,100)
plt.hist(dt1,bins=15,alpha=0.6,label='mean 0',edgecolor='white')
plt.hist(dt2,bins=15,alpha=0.6,label='mean 2',edgecolor='white')
plt.legend()
plt.title('Two overlapping histograms')
plt.show()
```

###### dt1=np.random.normal(0,1,100) means that an array named dt1 is created with 100 random numbers drawn from a normal (Gaussian) distribution that has a mean of 0 and a standard deviation of 1.
